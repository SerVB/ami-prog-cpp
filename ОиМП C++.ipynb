```c++
struct Point {
    int x, y;
};

bool operator == (const Point& p1, const Point& p2) {
    return p1.x == p2.x && p1.y == p2.y;
}

bool operator < (const Point& p1, const Point& p2) {
    return p1.x < p2.x || (p1.x == p2.x && p1.y < p2.y);
}

double distance(const Point& p) {
    return sqrt(p.x * p.x + p.y * p.y);
}

int main() {
    Point p{1, 2};
    p.x = 0;
    cout << p.y << "\n";
    
    vector<Point> points = {
        {1, 2},
        {9, -1}
    };
    points.push_back(Point{1, 2});
}
```

### Карта
```c++
#include <map>

int main() {
    std::map<char, int> freqs;
    std::string s;
    getline(std::cin, s);
    
    for (char c : s) {
        ++freqs[c];
    }
    
    auto it = freqs.find('a');
    if (it == freqs.end()) {
        std::cout << "Not found\n";
    } else {
        std::cout << "Found: " << it->first << " " << it->second << "\n";
    }
    
    for (const auto& item : freqs) {
        std::cout << item.first << " "
                << item.second << "\n";
    }
    
    /* Создать вектор из карты: */
    std::vector<std::pair<char, int>> v(freqs.begin(), freqs.end());
    std::sort(
        v.begin(),
        v.end(),
        [](const auto& a, const auto& b) {
            return a.second > b.second ||
                (a.second == b.second && a.first < b.first)
        }
    );
                
}
```

### Статические переменные являются общими для всех объектов этого класса
```c++
class C {
 public:
    static int counter;
    int id;
    C() {
        ++counter;
        id = counter;
        cout << "Constructor: " << counter << "\n";
    }
    
    C(const C& other) {
        ++counter;
        id = counter;
        cout << "Copy constructor: " << other.id << " " << id << "\n";
    }
    
    ~C() {
        cout << "Destructor: " << id << "\n";
    }
};

int C::counter = 0;

int main() {
    C object1, object2;  // Уничтожаются в обратном порядке (как извлечение из стека)
}
```

### Свой `back_inserter`
```c++
template <typename Container>
class BackInsertIterator {
 private:
    Container& cont;

 public:
    BackInsertIterator(Container& c) : cont(c) {}
    
    bool operator == (const BackInsertIterator& other) const {
        return true;
    }
    
    bool operator != (const BackInsertIterator& other) const {
        return !(*this == other);
    }
    
    typename Container::value_type operator * () {
        cont.push_back(ValueType());
        return cont.back();
    }
    
    BackInsertIterator& operator ++ () { // Префиксный, т.к. в параметрах ничего
        return *this;
    }
    
    BackInsertIterator operator ++ (int) { // Постфиксный, т.к. в параметрах фиктивный int
        auto temp = *this;
        ++*this;
        return temp;
    }
};

template <typename Container>
BackInsertIterator<Container> BackInserter(Container& c) {
    return {c};
}

template <typename InputIt, typename OutputIt>
OutputIt Copy(InputIt first, InputIt last, OutputIt dest) {
    while (first != last) {
        *dest++ = *first++;
    }
    return dest;
}
```

### Стандартный вектор
Не надо использовать этот самописный вектор вместо `std::vector`! Пишем вектор исключительно в *учебных* целях.
```c++
template <typename T>
struct VectorBase {  // RAII
    T* data;
    size_t cp;  // Capacity
    
    VectorBase(size_t n) {
        data = static_cast<T*>(operator new(n * sizeof(T)));  // Выделение сырых байтов
        cp = n;
    }
    
    ~VectorBase() {
        operator delete(data);
    }
                   
    void swap(VectorBase& other) {
        std::swap(data, other.data);
        std::swap(cp, other.cp);
    }
}

template <typename T>
class Vector {
private:
    VectorBase<T> base;
    size_t sz;  // Size
    
    void reallocate(size_t n) {
        VectorBase<T> tmp(n);
        size_t i;
        try {
            for (i = 0; i != n; ++i) {
                new (tmp.data + i) T(base.data[i]);
            }
        } catch (...) {
            for (size_t j = 0; j != i; ++j) {
                tmp.data[j].~T();
            }
            throw;
        }
        base.swap(tmp);
    }
public:
    Vector(size_t n, const T& val) : base(n) {
        size_t i;
        try {
            for (i = 0; i != n; ++i) {
                new (base.data + i) T(val);
            }
        } catch (...) {
            for (size_t j = 0; j != i; ++j) {
                base.data[j].~T();
            }
            throw;
        }
        sz = n;
    }
    
    ~Vector() {
        for (size_t i = 0; i != sz; ++i) {
            base.data[i].~T();
        }
    }
    
    void resize(size_t n) {
        if (n < sz) {
            for (size_t i = n; i != sz; ++i) {
                data[i].~T();
            }
            return;
        } else (n > sz) {
            // Не успели рассмотреть до конца
            if (n > cp) {
                reallocate(n);
            }
        }
    }
}
```
> Считается, что если ты делаешь `pop_back()` для пустого вектора, то ты сам себе Буратино.

&nbsp;
> Конструктор может кидать исключение, но деструктор не должен. Вы приходите в библиотеку за книгой, но ее может не оказаться (исключительная ситуация). А если вы приходите сдавать книгу, у вас ее обязательно возьмут (исключений нет)!
